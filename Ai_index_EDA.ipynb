```python
from sklearn import  model_selection, linear_model
import numpy as np
import pandas as pd
```

```python
x = np.array(df["Research"]).reshape(-1,1)
y = np.array(df["Total score"]).reshape(-1,1)

x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,test_size=0.2)

linear= linear_model.LinearRegression()

linear.fit(x_train,y_train)
```

```python
print("slope",linear.coef_)
print("Y- intercept",linear.intercept_)
```

```python
print("score",linear.score(x_train,y_train))
```

```python
predictions= linear.predict(x_test)

for x in range(len(predictions)):
  print("predicted value:",predictions[x], "Actual value:",y_test[x])
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,predictions,color="red")
plt.show()
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_test,predictions,color="red")

plt.axline(xy1=(0, float(linear.intercept_)), slope=float(linear.coef_))

plt.xlabel("Research")
plt.ylabel("Total score")
plt.title("Research vs Total score")
plt.show()
```
